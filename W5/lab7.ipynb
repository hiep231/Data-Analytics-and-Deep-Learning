```python
import nltk
from nltk.corpus import gutenberg
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.corpus import stopwords, wordnet
```

```python
# Tải xuống dữ liệu cần thiết
nltk.download('gutenberg')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# 1. Viết chương trình Python với thư viện NLTK để liệt kê các tên của copus.

# Lấy văn bản mẫu từ corpus Gutenberg
macbeth = gutenberg.raw('shakespeare-macbeth.txt')

# Tokenize văn bản
words = word_tokenize(macbeth)

# Gán nhãn từ loại
tagged_words = pos_tag(words)

# Lọc các từ là danh từ riêng (NNP, NNPS)
proper_nouns = {word for word, tag in tagged_words if tag in ['NNP', 'NNPS']}

# In danh sách các tên riêng
sorted(proper_nouns)
```

```python
# 2. Viết chương trình Python với thư viện NLTK để liệt kê danh sách các stopword bằng các ngôn ngữ khác nhau.
# Liệt kê các ngôn ngữ có sẵn trong NLTK
languages = stopwords.fileids()

# Hiển thị danh sách stopwords của từng ngôn ngữ
for lang in languages:
    print(f"\nStopwords của ngôn ngữ: {lang}")
    print(stopwords.words(lang)[:20])
```

```python
# 3. Viết chương trình Python với thư viện NLTK để kiểm tra danh sách các stopword bằng các ngôn ngữ khác nhau

languages = ['english', 'french', 'spanish']

for lang in languages:
    sw = set(stopwords.words(lang))
    print(f"\nNgôn ngữ: {lang}")
    print(f"Số lượng stopwords: {len(sw)}")
    print("10 stopwords đầu tiên:", list(sw)[:10])
```

```python
# 4. Viết chương trình Python với thư viện NLTK để loại bỏ các stopword từ một văn bản đã cho.
macbeth_filtered = [w for w in macbeth if w.lower() not in sw]
len(macbeth_filtered)
```

```python
# 5. Viết chương trình Python với thư viện NLTK bỏ qua các stopword từ danh sách các stopword.
from urllib import request
url = "http://www.gutenberg.org/files/2554/2554-0.txt"
response = request.urlopen(url)
raw = response.read().decode('utf8')
raw[:75]
```

```python
stop_words = set(stopwords.words('english'))
keep_words = {"not", "no", "but", "my"}
filtered_stopwords = stop_words - keep_words 

words = word_tokenize(raw)
filtered_words = [word for word in words if word.lower() not in filtered_stopwords]
print("Văn bản sau khi loại bỏ stopwords:", " ".join(filtered_words))
```

```python
# 6. Viết một chương trình Python với thư viện NLTK để tìm định nghĩa và ví dụ của một từ đã cho bằng WordNet từ Wikipedia
word = input("Nhập từ cần tra cứu: ")
synsets = wordnet.synsets(word)

for syn in synsets[:3]:  # Giới hạn 3 kết quả
    print(f"- Định nghĩa: {syn.definition()}")
    if syn.examples():
        print(f"  Ví dụ: {syn.examples()[0]}")
```

```python
# 7. Viết chương trình Python với thư viện NLTK để tìm tập hợp các từ đồng nghĩa và trái nghĩa của một từ nào đó
word = input("Nhập từ cần tìm đồng nghĩa và trái nghĩa: ")
synonyms = set()
antonyms = set()

for syn in wordnet.synsets(word):
    for lemma in syn.lemmas():
        synonyms.add(lemma.name())
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())

print(f"\n🔹 Từ đồng nghĩa của '{word}': {', '.join(synonyms) or 'Không có'}")
print(f"🔸 Từ trái nghĩa của '{word}': {', '.join(antonyms) or 'Không có'}")
```

```python
# 8. Viết chương trình Python với thư viện NLTK để có cái nhìn tổng quan về bộ tag, chi tiết của một tag cụ thể trong bộ tag và chi tiết về một số bộ tag liên quan, sử dụng biểu thức chính quy.
from nltk.data import load
import re

nltk.download('tagsets')
tagdict = load("help/tagsets/upenn_tagset.pickle")

print("Các tag trong NLTK:", ", ".join(list(tagdict.keys())[:10]), "...")

tag = input("Nhập tag (VD: NN, VB, JJ): ").upper()
print(tagdict.get(tag, "Không tìm thấy tag"))

pattern = input("Nhập regex (VD: NN.*): ")
print("🔹 Kết quả:", [t for t in tagdict if re.match(pattern, t)])
```

```python
# 9. Viết chương trình Python với thư viện NLTK để so sánh sự giống nhau của hai danh từ đã cho
nltk.download('wordnet')

noun1 = input("Nhập danh từ thứ nhất: ")
noun2 = input("Nhập danh từ thứ hai: ")

syn1 = wordnet.synsets(noun1, pos=wordnet.NOUN)
syn2 = wordnet.synsets(noun2, pos=wordnet.NOUN)

if syn1 and syn2:
    similarity = syn1[0].wup_similarity(syn2[0])
    print(f"Mức độ giống nhau giữa '{noun1}' và '{noun2}': {similarity:.2f}")
else:
    print("Không tìm thấy synset cho một trong hai từ.")
```

```python
# 10. Viết chương trình Python với thư viện NLTK để so sánh sự giống nhau của hai động từ đã cho.
verb1 = input("Nhập động từ thứ nhất: ")
verb2 = input("Nhập động từ thứ hai: ")

syn1 = wordnet.synsets(verb1, pos=wordnet.VERB)
syn2 = wordnet.synsets(verb2, pos=wordnet.VERB)

if syn1 and syn2:
    similarity = syn1[0].wup_similarity(syn2[0])
    print(f"Mức độ giống nhau giữa '{verb1}' và '{verb2}': {similarity:.2f}")
else:
    print("Không tìm thấy synset cho một trong hai từ.")
```

```python
# 11. Viết chương trình Python với thư viện NLTK để tìm số lượng tên nam và nữ trong các tên
# kho ngữ liệu. In tên 10 nam và nữ đầu tiên. Lưu ý: Kho văn bản tên chứa tổng cộng khoảng
# 2943 nam (male.txt) và 5001 nữ (Female.txt) tên. Kho được biên soạn bởi Kantrowitz, Ross.

import nltk
from nltk.corpus import names

nltk.download('names')

male_names = names.words('male.txt')
female_names = names.words('female.txt')

print(f"Số lượng tên nam: {len(male_names)}")
print(f"Số lượng tên nữ: {len(female_names)}")

print(f"10 tên nam đầu tiên: {male_names[:10]}")
print(f"10 tên nữ đầu tiên: {female_names[:10]}")
```

```python
# 12. Viết chương trình Python với thư viện NLTK để in 15 kết hợp ngẫu nhiên đầu tiên được gắn
# nhãn nam và được gắn nhãn tên nữ từ kho tên
import random

male_names = names.words('male.txt')
female_names = names.words('female.txt')

random_pairs = list(zip(random.sample(male_names, 15), random.sample(female_names, 15)))

print("15 cặp tên ngẫu nhiên:")
for i, (male, female) in enumerate(random_pairs, 1):
    print(f"{i}. 👨 {male} - 👩 {female}")
```

```python
# 13. Viết chương trình Python với thư viện NLTK để trích xuất ký tự cuối cùng của tất cả các tên
# được gắn nhãn và tạo mảng mới với chữ cái cuối cùng của mỗi tên và nhãn được liên kết.

male_names = names.words('male.txt')
female_names = names.words('female.txt')

name_endings = [(name[-1], 'male') for name in male_names] + [(name[-1], 'female') for name in female_names]

# In 10 phần tử đầu tiên
print("10 phần tử đầu tiên:")
print(name_endings[:10])
```
