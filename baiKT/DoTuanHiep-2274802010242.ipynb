```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Tạo DataFrame
data = {
    'Product': ['Laptop', 'Phone', 'Chair', 'Desk', 'Headphone', 'Monitor', 'Sofa', 'Keyboard', 'Mouse', 'Tablet'],
    'Category': ['Electronics', 'Electronics', 'Furniture', 'Furniture', 'Accessories', 'Electronics', 'Furniture', 'Accessories', 'Accessories', 'Electronics'],
    'Price': [1500, 800, 120, 300, 200, 600, 700, 150, 50, 400],
    'Date': ['2024-01-15', '2024-02-10', '2024-03-15', '2024-03-18', '2024-04-12', '2024-05-20', '2024-06-20', '2024-02-08', '2024-07-14', '2024-09-30'],
    'Quantity': [2, 5, 3, 1, 4, 2, 1, 6, 8, 3]
}
```

```python

df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# 2. Hiển thị thông tin DataFrame
df
```

```python
# 3. Giá trị trung bình của cột Price
df['Price'].mean()
```

```python
# 4. Nhóm dữ liệu và tính trung bình cột <Quantity>
avg_quantity = df.groupby('Category')['Quantity'].mean()
avg_quantity
```

```python
# 5. Sắp xếp DataFrame theo cột <Price> giảm dần
df_sorted = df.sort_values(by='Price', ascending=False)
df_sorted
```

```python
# 6. Vẽ biểu đồ cột cho <Price>
plt.bar(df['Product'], df['Price'], color='skyblue')
plt.title('Biểu đồ cột cho Price')
plt.xlabel('Product')
plt.ylabel('Price')
plt.show()
```

```python
plt.scatter(df['Quantity'], df['Price'], color='green')
plt.title('Scatter Plot for Quantity vs Price')
plt.xlabel('Quantity')
plt.ylabel('Price')
plt.show()
```

```python
# 8. Vẽ biểu đồ đường cho cột <Date>
df.plot(kind='line', x='Date', y='Price')
plt.title('Price Line Chart by Date')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
# 9. Vẽ biểu đồ cột thể hiện tổng doanh thu của từng <Category>
total_revenue = df.groupby('Category').apply(lambda x: (x['Quantity'] * x['Price']).sum())
plt.bar(total_revenue.index, total_revenue, color="skyblue")
plt.title('Total Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Total Revenue')
plt.show()

```

```python
# 10. Kiểm tra giá trị NaN
df.isna()
```

```python
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
sns.heatmap(df.isna(), 
            yticklabels=False, 
            cbar=True, 
            cmap='viridis')
```

```python
# 11. 11.Tìm các giá trị trùng lặp trong DataFrame dựa trên cột 'Product' và loại bỏ các hàng trùng lặp, giữ lại hàng đầu tiên.
df = df.drop_duplicates(subset='Product', keep='first')
df
```

```python
# 12. 12.Lọc ra các đơn hàng có số lượng (Quantity) lớn hơn 5
filtered_df = df[df['Quantity'] > 5]
filtered_df
```

```python
# 13. Tổng doanh thu
df['Revenue'] = df['Quantity'] * df['Price']
df
```

```python
# 14.Thêm một cột <Sales_Level> với điều kiện:
# + < 500  ->>  Low
# + 500 --- 2000  ->> Medium
# + >2000   ->> High
def sales_level(row):
    if row < 500:
        return 'Low'
    elif 500 <= row <= 2000:
        return 'Medium'
    else:
        return 'High'
```

```python

df['Sales_Level'] = df['Revenue'].apply(sales_level)
df
```

```python
# 15. Vẽ biểu đồ cột thể hiện số lượng đơn hàng ở từng mức <Sales_Level>.
plt.bar(df['Sales_Level'].value_counts().index, df['Sales_Level'].value_counts(), color="skyblue")
plt.title("Biểu đồ cột")
plt.xlabel("Danh mục")
plt.ylabel("Giá trị")
plt.show()
```

```python
# 16. Tạo một cột mới <Discount> với giá trị giảm giá 10% nếu doanh thu (Revenue) > 1000, ngược lại giảm giá 5%
def discount(row):
    if row > 1000:
        return 0.1
    else:
        return 0.05
```

```python
df['Discount'] = df['Revenue'].apply(discount)
df
```

```python
# 17. Tính doanh thu sau giảm giá <Final_Revenue> bằng công thức:
# Final_Revenue = Revenue - ( Revenue × Discount ).
df['Final_Revenue'] = df['Revenue'] - (df['Revenue'] * df['Discount'])
df
```

```python
# 18. Tạo cột <Month> lấy giá trị tháng từ cột <Date>.
df['Month'] = df['Date'].dt.month
df
```

```python
# 19. Vẽ biểu đồ đường thể hiện tổng doanh thu theo tháng.
total_revenue_month = df.groupby('Month')['Revenue'].sum()
plt.plot(total_revenue_month.index, total_revenue_month.values, label="Doanh thu theo tháng", color="blue", marker="o")
plt.title('Total Revenue by Month')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.show()

```
